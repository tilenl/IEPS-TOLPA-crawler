# PA2 context (in-repo)

Condensed reference for **Programming Assignment 2**: extract useful text from PA1 HTML, store **segments + embeddings** in **pgvector**, query and **rerank**.

Authoritative notes, rubric, and report checklist: `[PROGRAMMING-ASSIGNMENT-2.md](../PROGRAMMING-ASSIGNMENT-2.md)` in the repository root.

## Official assignment (three parts)

1. **Extract** meaningful content from HTML (XPath and/or regex); remove headers/menus/footers; prepare plain text; define **how** you split pages into thematic sections (paragraphs, headings, etc.) and justify in the report.
2. **Store** in a vector DB: extend PA1 PostgreSQL with **`cleaned_content`** on `page` and a **`page_segment`** table with **text + embedding**; enable `vector` extension; add **HNSW or IVFFlat** index.
3. **Retrieve**: embed the query, similarity search (cosine / L2 / inner product — experiment); demo program with **3 good** and **3 bad** queries; **cross-encoder reranking** on weak queries; document one case where reranking still fails.

Optional metadata on segments (heading, section tag, etc.) is encouraged if it helps retrieval — describe in the report.

## This project’s constraints (from our plan)

- **Data**: GitHub repository pages (image segmentation crawl). Target the **rendered README** region in HTML; layout is fairly uniform across repo pages.
- **Schema**: PA1 uses schema **`crawldb`** (not `crawl_db`). Migrations apply to `crawldb.page` and new `crawldb.page_segment`.
- **Embedding dimension**: `vector(D)` must match the chosen model (e.g. `all-MiniLM-L6-v2` → 384 — do not hard-code 768 unless the model outputs 768).
- **Lab code**: See `assignment-2-notebooks/vaje4_introduction_to_vector_databases.ipynb` and `assignment-2-notebooks/vaje4_vector_db_sample.ipynb` for pgvector + sentence-transformers patterns.
